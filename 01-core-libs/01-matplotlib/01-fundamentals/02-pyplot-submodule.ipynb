## 1. Understanding the pyplot Interface

`matplotlib.pyplot` is a collection of functions that make Matplotlib work like MATLAB. 
Each pyplot function makes some change to a figure: creating a figure, creating a plotting area, 
plotting lines, decorating the plot with labels, etc.

### Key Characteristics:

- **State-based**: pyplot tracks the current figure and plotting area
- **Implicit figure creation**: Creating plot elements automatically creates figures as needed
- **Simplified syntax**: Designed for ease of use and quick plotting
- **Interactive**: Well-suited for interactive environments like Jupyter notebooks

```python
# Import matplotlib.pyplot 
import matplotlib.pyplot as plt

# Check the pyplot module structure
print("Some key pyplot functions:")
functions = [func for func in dir(plt) if not func.startswith('_')][:30]  # First 30 non-private functions
for func in functions:
    print(f"- plt.{func}")
```

## 2. Pyplot's Stateful Programming Model

At the core of pyplot is its stateful nature. The module maintains an internal state that keeps 
track of the current figure, axes, and various settings. This is why you can create plots with 
minimal code but also why it can sometimes be confusing.


**Demonstrate the stateful nature of pyplot**

```python
# Create a figure without explicitly storing it
plt.figure(figsize=(10, 4))
plt.title("This is Figure 1")
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.show()
```

```python
# Create another figure - pyplot keeps track of the "current figure"
plt.figure(figsize=(10, 4))
plt.title("This is Figure 2")
plt.plot([1, 2, 3, 4], [1, 8, 27, 64], 'b^')
plt.show()
```

```python
# Return to the first figure using its automatically assigned number
plt.figure(1)
plt.title("Back to Figure 1 (modified)")
plt.xlabel("x values")
plt.show()
```

### The Current Figure and Axes

Pyplot functions operate on the "current" figure and axes. This is how pyplot maintains 
its state:

- `plt.figure()`: Creates a new figure or activates an existing one
- `plt.subplot()`: Creates a new subplot or activates an existing one
- `plt.gca()`: Gets the current axes
- `plt.gcf()`: Gets the current figure

```python
# Demonstrate accessing the current figure and axes
fig = plt.gcf()  # Get current figure
ax = plt.gca()   # Get current axes

print(f"Current figure number: {fig.number}")
print(f"Current axes position: {ax.get_position()}")
```

## 3. Key pyplot Functions

The pyplot module contains dozens of functions. Here are some of the most commonly used ones:

### Figure and Axes Creation
- `plt.figure()`: Create a new figure
- `plt.subplot()`: Create a subplot
- `plt.subplots()`: Create a figure and a set of subplots

### Plotting Functions
- `plt.plot()`: Plot lines and/or markers
- `plt.scatter()`: Create a scatter plot
- `plt.bar()`: Create a bar plot
- `plt.hist()`: Create a histogram
- `plt.imshow()`: Display an image

### Customization Functions
- `plt.title()`: Add a title
- `plt.xlabel()`, `plt.ylabel()`: Add axis labels
- `plt.legend()`: Add a legend
- `plt.grid()`: Add a grid
- `plt.colorbar()`: Add a colorbar

### Control Functions
- `plt.savefig()`: Save the current figure
- `plt.show()`: Display the figure
- `plt.close()`: Close a figure window

```python
# Create a simple demonstration of key pyplot functions
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))

# Plot data
plt.plot(x, y, 'b-', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')

# Add customizations
plt.title('Sine and Cosine Functions')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='best')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)  # Add a horizontal line at y=0

# Control the display
plt.tight_layout()
plt.show()
```

## 4. Practical Examples


Let's look at a few practical examples of using pyplot for common visualization tasks.

```python
# Example 1: Multiple plot types in a single figure
plt.figure(figsize=(12, 8))

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # row, column, index
plt.plot(x, np.sin(x))
plt.title('Line Plot: sin(x)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.scatter(x[::3], np.cos(x[::3]), c=x[::3], cmap='viridis')
plt.title('Scatter Plot: cos(x)')
plt.colorbar(label='x value')

plt.subplot(2, 2, 3)
plt.bar(range(10), np.random.rand(10))
plt.title('Bar Chart')

plt.subplot(2, 2, 4)
plt.hist(np.random.normal(0, 1, 1000), bins=30)
plt.title('Histogram')

plt.tight_layout()
plt.show()
```

```python
# Example 2: Controlling styles and colors with pyplot
plt.figure(figsize=(10, 6))

# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + np.pi/4)
y3 = np.sin(x + np.pi/2)

# Use pyplot's style system
plt.style.use('seaborn-v0_8-darkgrid')

# Create the plot with different colors and styles
plt.plot(x, y1, 'b-', linewidth=2, label='sin(x)')
plt.plot(x, y2, 'r--', linewidth=2, label='sin(x + Ï/4)')
plt.plot(x, y3, 'g:', linewidth=2, label='sin(x + Ï/2)')

# Add shaded regions
plt.fill_between(x, y1, 0, where=(y1 > 0), alpha=0.2, color='blue')
plt.fill_between(x, y1, 0, where=(y1 < 0), alpha=0.2, color='red')

# Add annotations
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2 - 1, 1.3),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Customize the plot
plt.title('Multiple Sine Waves with Style')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

# Set axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

plt.tight_layout()
plt.show()
```

## 5. Best Practices and Limitations

%% [markdown]
### Best Practices for pyplot

- **Consider your use case**: pyplot is ideal for interactive exploratory analysis
- **Clean up**: Use `plt.close()` to close figures when done (especially important for loops)
- **Be explicit about figure references** for complex visualizations
- **Use context managers** with `with plt.figure() as fig:` for better resource management
- **Set style settings** at the beginning of your script or notebook

### Limitations of pyplot

- **Memory management**: Can be inefficient if many figures are created but not closed
- **State-based confusion**: The "current figure" can sometimes lead to unexpected results
- **Complex layouts**: Can become unwieldy for highly customized visualizations
- **Threading issues**: Not thread-safe for complex applications
- **Less object-oriented**: Less flexibility compared to the OO interface

```python
# Demonstrate a common pyplot pitfall and its solution

# PITFALL: Creating plots in a loop without closing figures
plt.figure(figsize=(8, 2))
plt.title("This is the starting figure")
plt.plot([0, 1], [0, 1])
plt.show()

# Create several plots without closing them (comment out to avoid creating many figures)
"""
for i in range(3):
    plt.plot([0, 1], [0, i])
    plt.title(f"Plot {i}")
    plt.show()  # This doesn't close the figure!
"""

# SOLUTION: Properly close figures or use a context manager
for i in range(3):
    fig = plt.figure(figsize=(8, 2))
    plt.plot([0, 1], [0, i])
    plt.title(f"Plot {i} (with proper closing)")
    plt.show()
    plt.close(fig)  # Explicitly close the figure
```
